# ggplot2

Grammar of Graphics

Elements:


*   Data -მონეცემები რისი ვუზალიზაციაც გვსურს
 
*   Aesthetics - x-axis,y-axis,fill,col,shape,size.გრაფიკის ვიზუალური ელემენტი, ყველფერი რასაც ვზედავ.მგ. ერთი-ერთი სვეტი იყო x ღერძი 

*   Geometry - რა ტიპის გრაფიკი გვსურს მგ: ხაზოვანი დიაგრამა, სკატერფლოტი და ა.შ
*   Facet -  რა ნიშნითაც ვგსურს  მთლიანი მონაცემების დაყოფა.მგ: ქალი/კაცი, რეგიონი, ბანკი და ა.შ











```r
install.packages("tidyverse")
library(tidyverse)
```

```r
df=read.csv("/content/tidy_df.csv")
head(df)
```

##Scatter Plot

```r
ggplot(data=df) # ავარჩიოთ მონაცემები
#რადგან ჯერ მხოლოდ მონაცემები გვაქვს არჩეული და არ მიგვითითებია რა ტიპის ვიზუალიზაცია გვსურს, გამოგვიტანა ცარიელი გრაფიკი
```

```r
ggplot(data=df, aes(x=CIT_1, y=RoU)) #დავამატეთ Aesthetics
#რადგან ჯერ არ მიგვითითებია Geometry, ნახაზი არ არის სრული
```

```r
ggplot(data=df, aes(x=CIT_1, y=RoU))+geom_point()# დავამატეთ Geometry და უკვე გვაქვს საბოლოო სურათი
```

```r
#დავამატოთ Aesthetics-ში ფერები ბანკების მიხედვით
ggplot(data=df, aes(x=CIT_1, y=RoU,col=Bank))+geom_point()
```

```r
#დავამატოთ Aesthetics-ში დამატებით ფორმა ბანკების მიხედვით
ggplot(data=df, aes(x=CIT_1, y=RoU,col=Bank,shape=Bank))+geom_point()
```

```r
#დავამატოთ Aesthetics-ში დამატებით ზომა ass-ის მიხედვით
df %>% 
ggplot(aes(x=CIT_1, y=RoU,col=Bank,shape=Bank,size=ass, alpha=.5))+ #alpha არგუმენტით გავზარდოთ გამჭირველობა
geom_point()
```

# Histogram
ჰისტოგრამისათვის გამოიყენება geometry - `geom_histogram()`
შემავალი მონაცემები არის ერთ განზომილებიანი

```r
ggplot(data=df, aes(x=CIT_1))+
geom_histogram() 
```

```r
#შევცვალოთ bins მნიშვნელობა და დამატებით შევხედოთ რამდენად გასხვავებული განაწილება არის ბანკების მიხედვით CIT_1-ში. 
#ამისათვის aes-ში დავამატოთ fill=Bank როგორც წინა შემთხევაში
ggplot(data=df, aes(x=CIT_1,fill=Bank))+geom_histogram(bins=50, alpha = .59) 
```

```r
ggplot(data=df, aes(x=CIT_1,col=Bank,fill=Bank))+geom_histogram(bins=50) 
```

## Bar plot
Bar plot-თვის გამოიყენება geometry - `geom_bar()`
შემავალი მონაცემები არის ერთგანზომილებიანი, და ის არის კატეგორიული ცვლადი

```r
df %>% ggplot(aes(x=Tech,fill=Tech))+geom_bar() 
```

```r
ggplot(data=df, aes(x=Tech,fill=Bank))+geom_bar() 
```

```r
df %>% group_by(Bank,Tech) %>% 
  summarise(n = n()) %>%
  ggplot(aes(Bank, n, fill=Tech))+
  geom_bar(stat = 'identity', position = 'fill')+
  scale_y_continuous(labels = scales::percent)+
  ylab('Percent %')
```

##Frequency Polygon
Frequency Polygon-თვის გამოიყენება geometry - `geom_freqpoly()`
შემავალი მონაცემები არის ერთგანზომილებიანი, და ის არის უწყვეტი ცვლადი. გვეხმარება წარმოვიდგინოთ უწყვეტი ცვლადის განაწილება

```r
ggplot(data=df, aes(x=CIT_1,col=Bank))+geom_freqpoly()
```

# Density Plot
Density-ისათვის გამოიყენება geometry - `geom_density()` X ღერძზე არის რაოდენობრივი ცვლადი 

```r
df %>% ggplot(aes(RoU))+
geom_density(aes(fill=factor(Bank)), alpha=0.8) + 
    labs(title="Density plot", #ჩარტის სათაური
         subtitle="RoU Grouped by Bank", #ჩარტის ქვესათაური
         caption="Source: Sector_all",  
         x="RoU",
         fill="Banks")
```

#Boxplot & Violinplot
Boxplot-თვის გამოიყენება geometry - `geom_boxplot()`
violin-თვის გამოიყენება geometry - `geom_violin()`
x ღერძზე არის კატეგორიული ცვლადი და y ღერძზე არის უწყვეტი ცვალდი და გვიჩვენებს თითოეული კატეგორიისთვის უწყვეტი ცვლადის განაწილებას

```r
install.packages('cowplot')
library(cowplot)
p1 = ggplot(data=df, aes(x=Bank,y=CIT_1,fill=Bank))+geom_boxplot()+
p2 = ggplot(data=df, aes(x=Bank,y=CIT_1,fill=Bank))+geom_violin()
plot_grid(p1, p2, labels = "AUTO")
```

```r
#ჩავშალოთ თითოეულ ბანკში რეგიონების მიხედვით განაწილება
ggplot(data=df, aes(x=Bank,y=CIT_1,fill=Region))+geom_boxplot()
```

#Smooth line
smooth line-თვის გამოიყენება geometry - `geom_smooth()`
შემავალი ცვლადები არის ორგანზომილებიაი უწყვეტი ცვალდები. smooth line 
გვეხმარება ორი ცვალდის ერთმანეთზე დამოკიდებულება დავადგინოთ.



```r
ggplot(data=df, aes(x=Roe,y=CIT_1,col=Region))+ geom_smooth(se=F)# სტანდარტული გადახრა მოვაცილეთ se=F არგუმენტით
```

```r
#გავავლოთ რეგრესიის ხაზი method=lm ფუნქციის დახმარებით
ggplot(data=df, aes(x=Roe,y=CIT_1,col=Region))+ geom_smooth(method=lm,se=F)
```

```r
# შესაძლებელია ორი სურათის ერთამენთზე დამატება/გაერთიანება "+" სიმბოლოს გამოყენებით
#scatter plot და რეგრესიული ხაზი დავიტანოთ ერთ სურათზე
ggplot(data=df, aes(x=Roe,y=CIT_1,col=Region))+ geom_smooth(method=lm,se=F)+geom_point()
```

##Faceting
Faceting გვეხმარება მონაცემების გაყოფაში  რაიმე მახასიათებლის მიხედვით.
გამოიყენება ფუნქცია `facet_grid(~name_of_subgroup)`

```r
ggplot(data=df, aes(x=Roe,y=CIT_1,col=Region))+ geom_smooth(method=lm,se=F)+geom_point()+facet_grid(~Region)
```

ჩვენ შეგვიძლია ზემოთ აღნიშნული გრაფიკი შევინახოთ და ნელნელა დავამატოთ რამდენიმე ელემენტი:სათაური, ცვლადების სახელები, გავადიდოთ სურათი

```r
#შევინახოთ როგორც ობიექტი g
g=ggplot(data=df, aes(x=Roe,y=CIT_1,col=Region))+ geom_smooth(method=lm,se=F)+geom_point()+facet_grid(~Region)
```

## Title, names, zooming

*   გრაფიკის სათაური`ggtitle("name_of_plot")`
*   ღერძებისთვის სახელების დარქმევა `xlab("name_of_xaxis") + ylab("name_of_yaxis")`
*  მოვაცილოთ ცვალდების მნიშვნელობა გარკვეული შუალედისთვის scale_x_contious()
*  xlim() და ylim() ასევე არის ცვლადების გარკვეული მნიშვნელობების გაქრობა,
შესაბამისად იცვლება ცვლადების განაწილება (რეალურად არ ვაკეთებთ გადიდებას გრაფიკის)

*  coord_cartesian() რეალურად ცვლადების მნიშვნელობა არ ქრება, უბრალოდ გრაფიკი დიდდება. 




```r
g+ggtitle("Linear relationship between CIT_1 and ROE")+xlab("ROE")+ylab("CIT_1")
```

```r
g+ggtitle("Linear relationship between CIT_1 and ROE")+xlab("ROE")+ylab("CIT_1")+scale_x_continuous(limits = c(4, 5))
# როგორც ვხედავთ რამდენიმე მნიშვნელობა გაქრა
```

```r
g+ggtitle("Linear relationship between CIT_1 and ROE")+xlab("ROE")+ylab("CIT_1")+coord_cartesian(xlim = c(4, 5))
# გრაფიკი უბრალოდ გადიდდა, მნიშვნელობები არ გამქრალა შუალედის მიღმა
```
